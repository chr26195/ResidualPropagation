```python
import argparse
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'serif'
plt.style.use('bmh')

file_name = 'figs/csbm_GCN.pkl'
homo_p = np.arange(1, -0.01, -0.04)

color_map = plt.cm.get_cmap('twilight_shifted')
fig, axes = plt.subplots(nrows=4, sharex=True, figsize=(4, 4.5))
log = pd.read_pickle(file_name)

align_ntk_a_arrary = []
align_ntk_y_arrary = []
align_a_y_arrary = []

for i, p in enumerate(homo_p):
    k = 5
    c = color_map(1-(i+k)/(len(homo_p)+2*k)-0.04)
    axes[0].plot(log['steps'].values, log[f'test_acc_{p}'].values, linestyle='-', color=c, label=p, alpha=0.9)
    axes[1].plot(log['steps'].values, log[f'align_ntk_a_{p}'].values, linestyle='-', color=c, label=p, alpha=0.9)
    axes[2].plot(log['steps'].values, log[f'align_a_y_{p}'].values, linestyle='-', color=c, label=p, alpha=0.9)
    axes[3].plot(log['steps'].values, log[f'align_ntk_y_{p}'].values, linestyle='-', color=c, label=p, alpha=0.9)

for i, p in enumerate(homo_p):
    align_ntk_a_arrary.append(log[f'align_ntk_a_{p}'].values)
    align_ntk_y_arrary.append(log[f'align_ntk_y_{p}'].values)
    align_a_y_arrary.append(log[f'align_a_y_{p}'].values)

align_ntk_a_arrary = np.array(align_ntk_a_arrary)
align_ntk_y_arrary = np.array(align_ntk_y_arrary)
align_a_y_arrary = np.array(align_a_y_arrary)
axes[1].set_ylim(0, align_ntk_a_arrary.max() + 0.03)
axes[3].set_ylim(0, align_ntk_y_arrary.max() + 0.03)
axes[2].set_ylim(0, align_a_y_arrary.max() + 0.03)

for ax in axes:
    ax.grid(True)

axes[0].set_title('(a) Synthetic Dataset')
axes[0].set_ylabel('Test Acc.', fontsize=12)
axes[1].set_ylabel('$A$($\Theta_t$,A)', color='purple', fontsize=12)
axes[3].set_ylabel('$A$($\Theta_t$,$\Theta^*$)', color='royalblue', fontsize=12)
axes[2].set_ylabel('$A$(A,$\Theta^*$)', color='firebrick', fontsize=12)
axes[3].set_xlabel('Steps', fontsize=14)

axes[0].set_facecolor('snow')
axes[1].set_facecolor('snow')
axes[2].set_facecolor('snow')
axes[3].set_facecolor('snow')


plt.gcf().align_ylabels(axes)
plt.savefig('figs/alignment_csbm.pdf', bbox_inches = "tight", facecolor='white')
plt.show()
```

```python
import argparse
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'serif'
plt.style.use('bmh')

file_name = 'figs/cora_GCN_0.0.pkl'
drop_p = [.99, .9, .8, .7, .6, .5, .4, .3, .2, .1, 0]

color_map = plt.cm.get_cmap('twilight_shifted')
fig, axes = plt.subplots(nrows=3, sharex=True, figsize=(4, 4.5))
log = pd.read_pickle(file_name)

align_ntk_a_arrary = []
align_ntk_y_arrary = []

for i, p in enumerate(drop_p):
    k = 2
    c = color_map((i+k)/(len(drop_p)+2*k))
    axes[0].plot(log['steps'].values, log[f'test_acc_{p}'].values, linestyle='-', color=c, label=p, zorder=100+i*10, alpha=0.9)
    
    axes[1].plot(log['steps'].values, log[f'align_ntk_a_{p}'].values, linestyle='-', color=c, label=p, zorder=100+i*10, alpha=0.9)
    
    axes[2].plot(log['steps'].values, log[f'align_ntk_y_{p}'].values, linestyle='-', color=c, label=p, zorder=100+i*10, alpha=0.9)
    print((i+k)/(len(drop_p)+2*k))

for i, p in enumerate(drop_p):
    align_ntk_a_arrary.append(log[f'align_ntk_a_{p}'].values)
    align_ntk_y_arrary.append(log[f'align_ntk_y_{p}'].values)

align_ntk_a_arrary = np.array(align_ntk_a_arrary)
align_ntk_y_arrary = np.array(align_ntk_y_arrary)
axes[1].set_ylim(0, max(align_ntk_a_arrary.max(),log['align_ay'].values[0]) + 0.01)
axes[2].set_ylim(0, max(align_ntk_y_arrary.max(),log['align_ay'].values[0]) + 0.01)

for ax in axes:
    ax.grid(True)

axes[0].set_ylabel('Test Acc.', fontsize=14)
axes[0].set_title('(b) Real Homophilic Dataset')
axes[1].set_ylabel('$A$($\Theta_t$,A)', color='purple', fontsize=14)
axes[2].set_ylabel('$A$($\Theta_t$,$\Theta^*$)', color='royalblue', fontsize=14)

axes[2].set_xlabel('Steps', fontsize=14)

plt.gcf().align_ylabels(axes)

axes[0].set_facecolor('ghostwhite')
axes[1].set_facecolor('ghostwhite')
axes[2].set_facecolor('ghostwhite')

plt.savefig('figs/alignment_cora.pdf', bbox_inches = "tight", facecolor='white')
plt.show()
```
